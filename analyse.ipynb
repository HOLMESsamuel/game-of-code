```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
events = pd.read_csv("events_augmented.csv")
events.head()
```

```python
events_numerical_value = events[["attribute_count", "from_count", "host_count", "Tag"]]
events_numerical_value.head()
```

```python
corr = events_numerical_value.corr()
sns.heatmap(corr)
```

```python
none_events = events.loc[events["Tag"] == 0]
phishing_events = events.loc[events["Tag"] == 1]

plt.plot(none_events[["host_count"]], none_events[["attribute_count"]], 'bo')
plt.plot(phishing_events[["host_count"]], phishing_events[["attribute_count"]], 'ro')
```

```python
sns.pairplot(events_numerical_value)
```
